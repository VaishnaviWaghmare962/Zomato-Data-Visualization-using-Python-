# Which mode receives maximum rating

```python
plt.figure(figsize=(6,6))
chart=sns.boxplot(x="online_order",y="rate",data=df,palette="Paired")
chart.set_title("Boxplot of maximum rating ",fontsize=14,color="brown",fontweight="bold")
```

## conclusion : Maximum number of customer ordered online food.ofline order received lower rating in comparison to online order

```python
pivot_table=df.pivot_table(index="listed_in(type)",columns="online_order",aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,fmt='d',cmap="YlGnBu")
plt.title("Heatmap",color="darkblue",size=20)
plt.xlabel("Online Order")
plt.ylabel("Listed_in(type)")
plt.show()
```

## conclusion: Dining resturant primarily accept offline orders,whereas cafes primarily receive online order. 
## This suggests that client likes ordered at resturant,but perfer online ordering at cafes
